```python
from  PIL import Image # for opening images in python

import cv2 #for manipulating images in python
 
import pytesseract #a python library for google's tesseract model that does OCR

import pytesseract

pytesseract.pytesseract.tesseract_cmd = r"D:\volume E\temp programs\tesseract\tesseract.exe"



from matplotlib import pyplot as plt
%matplotlib inline
```

```python
#opening the images - loading images to memory
image1 = Image.open("dedication.jpg")
image2 = Image.open("dh.jpg")


```

```python
ocr_result1 = pytesseract.image_to_string(image1)
print(ocr_result1)
```

```python
ocr_result2 = pytesseract.image_to_string(image2)
print(ocr_result2)
```

```python
img = cv2.imread('dedication.jpg')


h, w, c = img.shape
boxes = pytesseract.image_to_boxes(image1) 
for b in boxes.splitlines():
    b = b.split(' ')
    img = cv2.rectangle(img, (int(b[1]), h - int(b[2])), (int(b[3]), h - int(b[4])), (0, 255, 0), 2)


plt.imshow(img)
plt.show()
```

```python
img = cv2.imread('dedication.jpg')

d = pytesseract.image_to_data(img, output_type=pytesseract.Output.DICT)

print(d.keys())
```

```python
n_boxes = len(d['text'])
for i in range(n_boxes):
    if int(d['conf'][i]) > 60:
        (x, y, w, h) = (d['left'][i], d['top'][i], d['width'][i], d['height'][i])
        img = cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)

        
plt.imshow(img)
plt.show()        
```

```python
img2 = cv2.imread('dh.jpg')
h2, w2, c2 = img2.shape

d2 = pytesseract.image_to_data(img2, output_type=pytesseract.Output.DICT)

print(d2.keys())
```

```python
n_boxes = len(d2['text'])
for i in range(n_boxes):
    if int(d2['conf'][i]) > 60:
        (x, y, w2, h2) = (d2['left'][i], d2['top'][i], d2['width'][i], d2['height'][i])
        img2 = cv2.rectangle(img2, (x, y), (x + w2, y + h2), (0, 255, 0), 2)

        
plt.imshow(img2)
plt.show()        
```

```python
# font 
font = cv2.FONT_HERSHEY_SIMPLEX 
  
# org 
org1 = (50, 50) 
org2 = (700, 50)  
# fontScale 
fontScale = 1
   
# Blue color in BGR 
color = (255, 0, 0) 
  
# Line thickness of 2 px 
thickness = 2
   
# Using cv2.putText() method 
img2 = cv2.putText(img2, 'OpenCV', org1, font,  
                   fontScale, color, thickness, cv2.LINE_AA) 
img2 = cv2.putText(img2, 'Humanities', org2, font,  
                   fontScale, color, thickness, cv2.LINE_AA) 
plt.imshow(img2)
plt.show()  
```
